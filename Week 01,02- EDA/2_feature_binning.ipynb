```python
df= pd.read_csv('data/processed/ChurnModelling_Missing_Values_Handled.csv')
```

```python
df.head(5)
```

```python
def custom_binning_credit_score(credit_score):
    if credit_score < 580:
        return 'poor'
    elif 580 <= credit_score < 670:
        return 'fair'
    elif 670 <= credit_score < 740:
        return 'good'
    elif 740 <= credit_score < 800:
        return 'very good'
    elif 800 <= credit_score <= 850:
        return 'excellent'
    else:
        assert True, "Credit score out of range"
```

```python
df['CreditScore_Binned'] = df['CreditScore'].apply(custom_binning_credit_score)
del df['CreditScore']
df.head(5)
```
